```python
# Cell 1: Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display plots inline
%matplotlib inline

# Cell 2: Load the dataset
# Make sure the file exists in data/raw/
df = pd.read_csv("../data/raw/Copy of Copy of data.csv")

# Quick preview
df.head()

# Cell 3: Dataset overview
print("Dataset shape:", df.shape)
print("\nColumns and data types:\n", df.dtypes)
print("\nMissing values per column:\n", df.isnull().sum())

# Cell 4: Summary statistics for numerical columns
df.describe()

# Cell 5: Summary statistics for categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns
df[categorical_cols].describe()

# Cell 6: Distribution of numerical features
numerical_cols = df.select_dtypes(include=['int64','float64']).columns
for col in numerical_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f'Distribution of {col}')
    plt.show()

# Cell 7: Distribution of categorical features
for col in categorical_cols:
    plt.figure(figsize=(6,4))
    sns.countplot(y=col, data=df)
    plt.title(f'Distribution of {col}')
    plt.show()

# Cell 8: Correlation heatmap
plt.figure(figsize=(10,8))
corr = df[numerical_cols].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()

# Cell 9: Outlier detection using boxplots
for col in numerical_cols:
    plt.figure(figsize=(6,4))
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot of {col}')
    plt.show()

# Cell 10: Top insights summary
print("Top 3-5 insights from EDA:")
print("1. Check for columns with missing values and decide on imputation strategy.")
print("2. Identify highly skewed numerical features that may require transformation.")
print("3. Detect outliers that may affect model training.")
print("4. Explore correlation between features to avoid multicollinearity.")
print("5. Review categorical feature distributions for rare categories.")

```
